```python
import os, json, requests
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()
client = OpenAI(api_key=os.environ["OPENAI_API_KEY"])
BASE = "https://nomad-movies.nomadcoders.workers.dev"
print("ready")
```

```python
def get_popular_movies():
    return requests.get(f"{BASE}/movies").json()

def get_movie_details(id):                                          
    return requests.get(f"{BASE}/movies/{id}").json()

def get_movie_credits(id):
    return requests.get(f"{BASE}/movies/{id}/credits").json()
print(get_movie_details(550))
```

```python
PROMPT_TEMPLATE = """
  You are a movie expert. You have access to the following functions:

  1. get_popular_movies()
     - Returns a list of currently popular movies.
     - Takes no arguments.

  2. get_movie_details(id)
     - Returns detailed information about a movie.
     - Argument: id (the movie id)

  3. get_movie_credits(id)
     - Returns the cast and crew of a movie.
     - Argument: id (the movie id)

  Please answer with the name of the function that you would like me to run.
  Please say nothing else, just the name of the function with the arguments.

  Answer the following question:

  {question}
  """

response = client.chat.completions.create(
      model="gpt-4o-mini",
      messages=[{"role": "user", "content": PROMPT_TEMPLATE}],
  )
print(response.choices[0].message.content)
```

```python
def ask(question):
      prompt = PROMPT_TEMPLATE.format(question=question)
      res = client.chat.completions.create(
          model="gpt-4o-mini",
          messages=[{"role": "user", "content": prompt}],
      )
      print(f"Q: {question}")
      print(f"A: {res.choices[0].message.content}\n")
```

```python
ask("지금 인기 있는 영화가 무엇인지 알려줘")
```

```python
ask("movie ID 550에 해당하는 영화가 무엇인지 알려줘")
```
